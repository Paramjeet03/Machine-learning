```python
import numpy as np
import pandas as pb
```

```python
df=pb.read_csv(r"C:\Users\Paramjeet\Desktop\MACHINE LEARNING\archive (4)\Salary Data.csv")
df
```

```python
df=df.dropna()
```

```python
feature=df[["Years of Experience"]].values
feature
```

```python
target=df.Salary
```

```python
from sklearn.preprocessing import PolynomialFeatures
pf=PolynomialFeatures(degree=3)
pf
```

```python
pff=pf.fit_transform(feature)
pff
```

```python
from sklearn.linear_model import LinearRegression
modal=LinearRegression()
modal
```

```python
modal.fit(pff,target)
```

```python
sample=pf.transform([[ 5. ]])

modal.predict(sample)
```
