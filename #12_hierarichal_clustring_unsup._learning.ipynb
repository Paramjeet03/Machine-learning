# steps In agglomerative :-

> 1) In which every point assum a cluster.

> 2) Make a new cluster by connecting nearest point.

> 3) Doing step 2 until points gets single cluster.


# steps In Divisive :-

> 1) In which assum a single cluster which having all the points.

> 2) separete the nearest point from cluster and form new saperate cluster.

> 3) Doing step 2 until points gets own cluster.

# Implentation :-

```python
import numpy as np
import pandas as pb
import matplotlib.pyplot as plt 
```

```python
df=pb.read_csv(r"C:\Users\Paramjeet\Desktop\MACHINE LEARNING\DATASETS\IRIS.csv")
df
```

```python
df=df.iloc[:,0:3]
```

```python
df.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scale=StandardScaler()
x_scale=scale.fit_transform(df)
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
```

```python
pca_df=pca.fit_transform(x_scale)
```

```python
pca_df.shape
```

```python
plt.scatter(pca_df[:,0],pca_df[:,1])
```

```python
# Agglomeritive clustring :-
## to construct dendogram
import scipy.cluster.hierarchy as sc
## plot dendogram
plt.figure(figsize=(10,5))
plt.title("Dendogram")
sc.dendrogram(sc.linkage(pca_df,method="ward"))
plt.title("Dendogram")
plt.xlabel("sample of index")
plt.ylabel("Ecludien distance")
plt.show()
```

## No. of cluster :- 2

```python
from sklearn.cluster import AgglomerativeClustering
modal=AgglomerativeClustering(n_clusters=2,affinity="euclidean",linkage="ward")
```

```python
modal.fit(pca_df)
```

```python
modal.labels_
```

```python
plt.scatter(pca_df[:,0],pca_df[:,1],c=modal.labels_)
plt.show()
```
