# Votting :-

> That use diff. type of modals (estimator) at a same time.

> This work on the concept of votting . The assign diff. value of vote to modals internally by which good output is gernated. 

# For ex :-

# modal         value of    Total       vote
                   vote        vote        count

#  phd     -->    1     -->   50   -->   50
#  
#  Btech   -->    0.5   -->   50   -->   25
#  
#  10th    -->    0.1   -->   50   -->    5

> not use the bootstrap method.

```python
import pandas as pb
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
df=pb.read_csv(r"A:\Ducat\PYTHON\phyton\data analytics\MACHINE LEARNING\DATASETS\IRIS.csv")
df
```

```python
x=df.iloc[:,0:4].values
x
```

```python
y=df.species.values
y
```

```python
from sklearn.preprocessing import MinMaxScaler
sc=MinMaxScaler()
x_sc=sc.fit_transform(x)
from sklearn.model_selection import train_test_split
x_train1,x_test1,y_train1,y_test1=train_test_split(x_sc,y,train_size=0.7,random_state=20)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import VotingClassifier
m1=LogisticRegression(max_iter=2000)
m2=DecisionTreeClassifier(max_depth=3)
m3=RandomForestClassifier(max_depth=2)
modal=VotingClassifier(estimators=[("1st modal",m1),("2nd modal",m2),("3rd modal",m3)])
```

```python
modal.fit(x_train1,y_train1)
```

```python
xtest_p=modal.predict(x_test1)
xtrain_p=modal.predict(x_train1)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print("Training score :- ",accuracy_score(y_train1,xtrain_p))
print("Testing score :- ",accuracy_score(y_test1,xtest_p))
```
