# Activation fuction :- These are use too  make more accurate the prediction . There are 2 type type of function use in classification regression.

## 1.) Sigmoid function (ovr):- it is best for only 2 class.

## 2.) Softmax function (multimonial):- it is best for more than 2 class.

```python
import numpy as np
import pandas as pb
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
df=pb.read_csv(r"A:\Ducat\PYTHON\phyton\data analytics\MACHINE LEARNING\DATASETS\IRIS.csv")
df
```

```python
df.species.value_counts()
```

```python
feature=df.iloc[:,:4].values
print("dimension :- ",feature.ndim,"\n",feature) #2d array
```

```python
target=df.species
```

```python
from sklearn.linear_model import LogisticRegression
```

## ---> Model by Auto (Work only some time) :-

```python
model1=LogisticRegression()
model1
```

```python
model1.fit(feature,target)
```

```python
model1.predict([[4.9,  3. , 1.4, 0.2]])
```

```python
model1.predict_proba([[4.9,  3. , 1.4, 0.2]])
```

## ---> Model by ovr (sgmoid) :-

```python
model=LogisticRegression(multi_class='ovr')
model
```

```python
model.fit(feature,target)
```

```python
model.predict([[4.9,  3. , 1.4, 0.2]])
```

```python
model.predict_proba([[4.9,  3. , 1.4, 0.2]])
```

## ---> Model by multinomial (softmax) :-

```python
model2=LogisticRegression(multi_class='multinomial')
model2
```

```python
model2.fit(feature,target)
```

```python
model2.predict([[4.9,  3. , 1.4, 0.2]])
```

```python
model2.predict_proba([[4.9,  3. , 1.4, 0.2]])
```
