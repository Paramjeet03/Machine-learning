# Implentation :-

```python
import numpy as np
import pandas as pb
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
x,y=make_blobs(n_samples=10000,n_features=2,centers=3,random_state=27)
```

```python
x.shape
```

```python
plt.scatter(x[:,0],x[:,1])
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=27)
```

```python
from sklearn.cluster import KMeans
```

```python
# finding the k value.
# mannual way 
# elbow method
```

```python
wcss=[]
for k in range(1,11):
    kmeans=KMeans(n_clusters=k,init='k-means++')
    kmeans.fit(x_train)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,11),wcss)
plt.xticks(range(1,11))
plt.xlabel("No. of centeroids")
plt.ylabel("Wcss")
plt.show()
# so no. of centroids is 3
```

```python
kmeans=KMeans(n_clusters=3,init="k-means++")
y_train_label=kmeans.fit_predict(x_train)
y_test_label=kmeans.predict(x_test)
```

# -> Before the clustring :-

```python
plt.scatter(x[:,0],x[:,1],c=y)
plt.show()
```

# After the clusting :-

```python
plt.scatter(x_train[:,0],x_train[:,1],c=y_train_label)
plt.show()
```

```python
plt.scatter(x_test[:,0],x_test[:,1],c=y_test_label)
plt.show()
```
